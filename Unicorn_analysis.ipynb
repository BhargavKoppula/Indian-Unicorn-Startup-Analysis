```python
def convert_to_number(x):
    if isinstance(x, str):
        x = x.replace(',', '').strip()
        match = re.match(r'[$?]?([\d\.]+)\s*(K|M|B|Cr|L)', x)
        if match:
            num = float(match.group(1))
            unit = match.group(2)

            multiplier = {
                'K': 1_000,
                'M': 1_000_000,
                'B': 1_000_000_000,
                'Cr': 10_000_000,
                'L': 100_000,
            }.get(unit, 1)

            return int(num * multiplier)

    return pd.NA
```

```python
df['Total_funding'] = df['total_funding_till_date'].apply(convert_to_number).astype('Int64')
```

```python
df['Annual_Revenue'] = df['annual_revenue_amount'].apply(convert_to_number).astype('Int64')
df['Valuation_Amount'] = df['valuation_amount'].apply(convert_to_number).astype('Int64')
```
